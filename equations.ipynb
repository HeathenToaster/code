```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
```

```python
def reward(alpha, gamma, mvtT):
    return alpha/(1+gamma * mvtT)
def effort(a, mass, mvtT, b, dist, gamma, i, j):
    return (a*mass*mvtT + ((b*mass*dist**i) / (mvtT**(j-1)))) / (1+(gamma*mvtT))
def effortwalk(a, mass, mvtT, b, dist, gamma):
    return (a*mass*mvtT + ((b*mass*dist**2) / (mvtT))) / (1+(gamma*mvtT))
```

```python
T = np.linspace(0, 1, 100)
```

$$
  Reward = \frac{\alpha} {1+\gamma*T}
$$
$$
  Effort = -\frac{amT + \frac{bmdÂ²}{T}}{1+\gamma*T}
$$

```python
 
def calc_parabola_vertex(x1, y1, x2, y2, x3, y3):
    denom = (x1-x2) * (x1-x3) * (x2-x3);
    A     = (x3 * (y2-y1) + x2 * (y1-y3) + x1 * (y3-y2)) / denom;
    B     = (x3*x3 * (y1-y2) + x2*x2 * (y3-y1) + x1*x1 * (y2-y3)) / denom;
    C     = (x2 * x3 * (x2-x3) * y1+x3 * x1 * (x3-x1) * y2+x1 * x2 * (x1-x2) * y3) / denom;
    return A,B,C


from ipywidgets import interactive

def v(Vmax, tf):
    plt.figure(1)
    x1,y1=[0,0]
    x2,y2=[tf,0]
    x3,y3=[tf/2,Vmax]

    a,b,c=calc_parabola_vertex(x1, y1, x2, y2, x3, y3)
    x_pos=np.linspace(0,tf,100)
    y_pos=[]

    #Calculate y values 
    for x in range(len(x_pos)):
        x_val=x_pos[x]
        y=(a*(x_val**2))+(b*x_val)+c
        y_pos.append(y)

    print("y=", a, "x**2 + ", b, "x +", c)
    plt.plot(x_pos, y_pos, linestyle='-.', color='black') # parabola line
    plt.scatter(x_pos, y_pos, color='gray') # parabola points
    plt.scatter(x1,y1,color='r',marker="D",s=50) # 1st known xy
    plt.scatter(x2,y2,color='g',marker="D",s=50) # 2nd known xy
    plt.scatter(x3,y3,color='k',marker="D",s=50) # 3rd known xy
    plt.xlim(0, 5)
    plt.ylim(0, 150)
    plt.show() 


interactive_plot = interactive(v, Vmax=(0, 100, 10), tf=(0, 5.0))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```
