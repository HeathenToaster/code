```python
# explore distributions

import warnings
import numpy as np
import pandas as pd
import scipy.stats as st
import statsmodels.api as sm
from scipy.stats._continuous_distns import _distn_names
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['figure.figsize'] = (16.0, 12.0)

BEST = []

# Create models from data
def best_fit_distribution(data, bins=200, ax=None):
    """Model data by finding best fit distribution to data"""
    # Get histogram of original data
    y, x = np.histogram(data, bins=bins, density=True)
    x = (x + np.roll(x, -1))[:-1] / 2.0

    # Best holders
    best_distributions = []

    # Estimate distribution parameters from data
    for ii, distribution in enumerate([d for d in _distn_names if not d in ['levy_stable', 'studentized_range'] + BEST]):

        # print("{:>3} / {:<3}: {}".format( ii+1, len(_distn_names), distribution ))

        distribution = getattr(st, distribution)

        # Try to fit the distribution
        try:
            # Ignore warnings from data that can't be fit
            with warnings.catch_warnings():
                warnings.filterwarnings('ignore')
                
                # fit dist to data
                # params = distribution.fit(data, floc=0, fscale=1)
                params = distribution.fit(data,)


                # Separate parts of parameters
                arg = params[:-2]
                loc = params[-2]
                scale = params[-1]

                
                # Calculate fitted PDF and error with fit in distribution
                pdf = distribution.pdf(x, loc=loc, scale=scale, *arg)
                # pdf = distribution.pdf(x,  *arg)
                # pdf = distribution.cdf(x, loc=loc, scale=scale, *arg)
                sse = np.sum(np.power(y - pdf, 2.0))
                
                # if axis pass in add to plot
                try:
                    if ax:
                        pd.Series(pdf, x).plot(ax=ax)
                    end
                except Exception:
                    pass

                # identify if this distribution is better
                best_distributions.append((distribution, params, sse))
        
        except Exception:
            pass

    
    return sorted(best_distributions, key=lambda x:x[2])

def make_pdf(dist, params, size=10000):
    """Generate distributions's Probability Distribution Function """

    # Separate parts of parameters
    arg = params[:-2]
    loc = params[-2]
    scale = params[-1]

    # Get sane start and end points of distribution
    start = dist.ppf(0.01, *arg, loc=loc, scale=scale) if arg else dist.ppf(0.01, loc=loc, scale=scale)
    end = dist.ppf(0.99, *arg, loc=loc, scale=scale) if arg else dist.ppf(0.99, loc=loc, scale=scale)

    # Build PDF and turn into pandas Series
    x = np.linspace(start, end, size)
    y = dist.pdf(x, loc=loc, scale=scale, *arg)
    pdf = pd.Series(y, x)

    return pdf

def make_cdf(dist, params, size=10000):
    """Generate distributions's Probability Distribution Function """

    # Separate parts of parameters
    arg = params[:-2]
    loc = params[-2]
    scale = params[-1]

    # Get sane start and end points of distribution
    start = dist.ppf(0.01, *arg, loc=loc, scale=scale) if arg else dist.ppf(0.01, loc=loc, scale=scale)
    end = dist.ppf(0.99, *arg, loc=loc, scale=scale) if arg else dist.ppf(0.99, loc=loc, scale=scale)

    # Build PDF and turn into pandas Series
    x = np.linspace(start, end, size)
    y = dist.cdf(x, loc=loc, scale=scale, *arg)
    cdf = pd.Series(y, x)

    return 1-cdf

data = pd.Series(wait_after_seqALL[animal]["1100"])
# data = np.log(pd.Series(wait_after_seqALL[animal]["1100"]))
# data = pd.Series(sorted({x for v in wait_after_seq[animal].values() for x in v}))

# Plot for comparison
plt.figure(figsize=(12,8))
ax = data.plot(kind='hist', bins=50, density=True, alpha=0.5, color=list(matplotlib.rcParams['axes.prop_cycle'])[1]['color'])
ax.set_ylim(0.01, .5)
ax.set_title(u'data.\n All Fitted Distributions')
ax.set_xlabel(u'wait')
ax.set_ylabel('Frequency')


best_distibutions = best_fit_distribution(data, 200, ax)
best_dist = best_distibutions[0:5]

plt.figure(figsize=(12,8))
for dist in best_dist:
    cdf = make_cdf(dist[0], dist[1])
    
    param_names = (dist[0].shapes + ', loc, scale').split(', ') if dist[0].shapes else ['loc', 'scale']
    param_str = ', '.join(['{}={:0.2f}'.format(k,v) for k,v in zip(param_names, dist[1])])
    dist_str = '{}({})'.format(dist[0].name, param_str)

    ax = cdf.plot(lw=2, label=f'{dist_str} CDF', legend=True)
    ax.hist(data, 
    # bins=np.logspace(np.log10(0.1),np.log10(301),300),
    density=True, cumulative=-1, alpha=0.5, histtype='step', label='data', color=list(matplotlib.rcParams['axes.prop_cycle'])[1]['color'])

    
    print(dist[0].name, param_str, dist[2])
    # ax.set_title(u'data. with best fit distribution \n' + dist_str)
    ax.set_xlabel(u'wait')
    ax.set_ylabel('Frequency')
    ax.set_xscale('log',)
    ax.set_yscale('log')
```

```python
from pyddm import Fittable, Fitted, Sample, Model, plot, set_N_cpus
from pyddm.models import DriftConstant, NoiseConstant, BoundConstant, OverlayNonDecision, ICPoint, LossRobustLikelihood
from pyddm.functions import fit_adjust_model, display_model, solve_partial_conditions

set_N_cpus(6)

def get_values_from_parameters(parameters):
    drift = parameters['drift']['drift'].default()
    # noise = parameters['noise']['noise'].default()
    noise = parameters['noise']['noise']
    # bound = parameters['bound']['B'].default()
    bound = parameters['bound']['B']
    IC = parameters['IC']['x0'].default()
    # nondectime = parameters['overlay']['nondectime'].default()
    nondectime = parameters['overlay']['nondectime']

    return drift, noise, bound, IC, nondectime

def DDMfit(array): # fix non-decision time and noise
    T_dur=np.max(array)+1
    dt=.01
    sample = Sample(np.asarray(array), np.asarray([]*len(array)), 0)
    model_fit = Model(name='Simple model (fitted)',
                  IC=ICPoint(x0=Fittable(minval=90, maxval=100)),
                  drift=DriftConstant(drift=Fittable(minval=0, maxval=5)),
                #   noise=NoiseConstant(noise=Fittable(minval=0, maxval=50)),
                  noise=NoiseConstant(noise=1),
                #   bound=BoundConstant(B=Fittable(minval=0, maxval=100)),
                  bound=BoundConstant(B=100),
                  overlay=OverlayNonDecision(nondectime=0),
                  dx=.01, dt=dt, T_dur=T_dur)

    fitted = fit_adjust_model(sample, model_fit,
                fitting_method="differential_evolution",
                lossfunction=LossRobustLikelihood,
                verbose=False);

    drift, noise, bound, IC, nondectime = get_values_from_parameters(model_fit.parameters())
    pdf_y = fitted.solve(conditions={"left_is_correct": 1}).pdf_corr()
    cdf_y = fitted.solve(conditions={"left_is_correct": 1}).cdf_corr()
    pdf_x = np.arange(0, T_dur+dt, dt)
    loss = fitted.fitresult.value()
    return pdf_x, pdf_y, cdf_y, drift, noise, bound - IC, nondectime, loss
    
def drawfromDDM(mean, std, A, t0, N):
    # waits=drawfromDDM(mean, std, A, t0, N)
    T_dur=10000
    dt=.01
    model = Model(name='generator',
                  IC=ICPoint(x0=100-A),
                  drift=DriftConstant(drift=mean),
                  noise=NoiseConstant(noise=std),
                  bound=BoundConstant(B=100),
                  overlay=OverlayNonDecision(nondectime=t0),
                  dx=.01, dt=dt, T_dur=T_dur)
    sol = model.solve()
    sample = sol.resample(N)
    df = sample.to_pandas_dataframe()
    return df['RT'].to_numpy()

test = [4.519999999999982, 1.7999999999999545, 19.879999999999995, 1.080000000000041, 2.6399999999999864, 4.439999999999998, 0.6000000000000227, 15.080000000000041,
        2.4399999999999977, 0.19999999999998863, 0.6000000000000227, 1.5199999999999818, 3.8000000000000114, 0.5199999999999818, 7.0, 13.240000000000009,
        0.7599999999999909, 2.240000000000009, 0.8000000000000682, 0.4399999999999977, 2.3600000000000136, 29.360000000000014, 0.5600000000000591]
        
pdf_x, pdf_y, cdf_y, drift, noise, bound, nondectime, loss = DDMfit(test)
print(f'drift: {drift:.2f}, noise: {noise:.2f}, bound: {bound:.2f}, nondectime: {nondectime:.2f}')
# plt.plot(cdf_x, 1-cdf_y)
# plt.xlim(0, 25)
```

```python
# maximum likelihood


def Wald_pdf(x, alpha, theta, gamma):
    x = x - theta
    x[x<0] = 1e-10
    arg = 2 * np.pi * x ** 3
    res = alpha / np.sqrt(arg) * np.exp(-((alpha-gamma * x) ** 2) / (2 * x))
    return np.array(res, dtype=np.float64)

from scipy.stats import norm
def Wald_cdf(x, alpha, theta, gamma):
    x = x - theta
    x[x<0] = 1e-10
    return np.array(norm.cdf((gamma*x-alpha)/np.sqrt(x)) + np.exp(2*alpha*gamma)*norm.cdf(-(gamma*x+alpha)/np.sqrt(x)), dtype=np.float64)

def log_lik_wald(x, params):
    alpha, theta, gamma = params
    pdf_vals = Wald_pdf(x, alpha, theta, gamma)
    ln_pdf_vals = np.log(pdf_vals)
    log_lik_val = ln_pdf_vals.sum()
    return log_lik_val

def crit(params, *args):
    alpha, theta, gamma = params
    x = args
    
    pdf_vals = Wald_pdf(x, alpha, theta, gamma)
    ln_pdf_vals = np.log(pdf_vals)
    log_lik_val = ln_pdf_vals.sum()
    neg_log_lik_val = -log_lik_val
    return neg_log_lik_val

def wald_fit(x, alpha_init=2, theta_init=0, gamma_init=.5):
    params_init = np.array([alpha_init, theta_init, gamma_init])
    res = scipy.optimize.minimize(crit, params_init, args=x, bounds=((0, None), (0, 1e-8), (0, None)))
    return res.x
    
waits=drawfromDDM(1, 1, 5, 0, 10000)
a, t, g = wald_fit(waits)
```
